```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df = pd.read_csv("../../evaluation/sgd/netflow/sgd_evaluation.csv")
df
```

```python
# Create plot for anomaly class metrics vs batch size
fig = go.Figure()

# Add traces for each metric
fig.add_trace(go.Scatter(
    x=df['batch_size'],
    y=df['precision_anomaly'],
    mode='lines+markers',
    name='Precision',
    line=dict(color='blue')
))

fig.add_trace(go.Scatter(
    x=df['batch_size'],
    y=df['recall_anomaly'],
    mode='lines+markers',
    name='Recall',
    line=dict(color='red')
))

fig.add_trace(go.Scatter(
    x=df['batch_size'],
    y=df['f1_anomaly'],
    mode='lines+markers',
    name='F1-Score',
    line=dict(color='green')
))

# Update layout
fig.update_layout(
    title='SGD Classifier Performance Metrics for Anomaly Class vs Batch Size',
    xaxis_title='Batch Size',
    yaxis_title='Score',
    yaxis_range=[0, 1],
    legend_title='Metric',
    template='plotly_white'
)

# Show the plot
fig.show()
```
