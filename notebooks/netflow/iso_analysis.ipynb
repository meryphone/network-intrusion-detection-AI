```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



# Crear figura
fig, ax = plt.subplots(figsize=(10, 6))

# Plotear las 3 líneas principales
ax.plot(df['contamination'], df['recall_normal'], 
        marker='o', linewidth=2.5, markersize=7, 
        label='Recall class 0 (Normal)', color='#1f77b4', alpha=0.8)

ax.plot(df['contamination'], df['recall_anomaly'], 
        marker='s', linewidth=2.5, markersize=7, 
        label='Precision class 1 (Anomaly)', color='#ff7f0e', alpha=0.8)

ax.plot(df['contamination'], df['f1_normal'], 
        marker='^', linewidth=2.5, markersize=7, 
        label='F1 class 0 (Normal)', color='#2ca02c', alpha=0.8)

# Formateo del gráfico
ax.set_xlabel('Contamination', fontsize=12, fontweight='bold')
ax.set_ylabel('Metric value', fontsize=12, fontweight='bold')
ax.set_title('Isolation Forest: metrics vs contamination (class 0)', 
             fontsize=13, fontweight='bold')
ax.legend(fontsize=10, loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_ylim([0.3, 1.05])
ax.set_xlim([0, 0.55])

# Estilo similar a la imagen
plt.tight_layout()
plt.savefig('isolation_forest_metrics.png', dpi=300, bbox_inches='tight')
plt.show()

print("Graph saved: isolation_forest_metrics.png")
```
