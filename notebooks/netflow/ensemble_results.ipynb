```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df = pd.read_csv("../../evaluation/ensemble/netflow/ensemble_evaluation.csv")
df
```

```python
# Create plot for anomaly class metrics vs batch size
fig = go.Figure()

# Add traces for each metric
fig.add_trace(go.Scatter(
    x=df['batch_size'],
    y=df['precision_anomaly'],
    mode='lines+markers',
    name='Precision',
    line=dict(color='blue')
))

fig.add_trace(go.Scatter(
    x=df['batch_size'],
    y=df['recall_anomaly'],
    mode='lines+markers',
    name='Recall',
    line=dict(color='red')
))

fig.add_trace(go.Scatter(
    x=df['batch_size'],
    y=df['f1_anomaly'],
    mode='lines+markers',
    name='F1-Score',
    line=dict(color='green')
))

# Update layout
fig.update_layout(
    title='Ensemble Classifier Performance Metrics for Anomaly Class vs Batch Size',
    xaxis_title='Batch Size',
    yaxis_title='Score',
    yaxis_range=[0, 1],
    legend_title='Metric',
    template='plotly_white'
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Cargar datos
df = pd.read_csv("../../evaluation/ensemble/netflow/ensemble_evaluation.csv")

# Crear figura
fig = go.Figure()

# Tiempo total vs Batch Size
fig.add_trace(go.Scatter(
    x=df["batch_size"],
    y=df["total_time_sec"],
    mode="lines+markers",
    name="Tiempo Total",
    text=df["batch_size"],
    hovertemplate="<b>Batch Size:</b> %{x}<br><b>Tiempo Total:</b> %{y:.2f}s<extra></extra>",
    marker=dict(size=12, color="blue"),
    line=dict(width=3)
))

# Tiempo de test vs Batch Size
fig.add_trace(go.Scatter(
    x=df["batch_size"],
    y=df["test_time_sec"],
    mode="lines+markers",
    name="Tiempo Test",
    text=df["batch_size"],
    hovertemplate="<b>Batch Size:</b> %{x}<br><b>Tiempo Test:</b> %{y:.2f}s<extra></extra>",
    marker=dict(size=12, color="green"),
    line=dict(width=3)
))

# Tiempo warmup vs Batch Size
fig.add_trace(go.Scatter(
    x=df["batch_size"],
    y=df["warmup_time_sec"],
    mode="lines+markers",
    name="Tiempo Warmup",
    text=df["batch_size"],
    hovertemplate="<b>Batch Size:</b> %{x}<br><b>Tiempo Warmup:</b> %{y:.2f}s<extra></extra>",
    marker=dict(size=12, color="red"),
    line=dict(width=3)
))

# Configurar layout
fig.update_layout(
    title="Tiempo de Procesamiento vs Tamaño de Batch",
    xaxis=dict(title="Batch Size"),
    yaxis=dict(title="Tiempo (segundos)"),
    hovermode="x unified",
    legend=dict(x=0.01, y=0.99),
    height=600,
    width=1000,
    font=dict(size=12),
    plot_bgcolor="rgba(240,240,240,0.5)"
)

fig.write_html("time_vs_batch.html")
fig.show()
```
