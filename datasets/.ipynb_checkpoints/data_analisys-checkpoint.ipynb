```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import RobustScaler


df_train= pd.read_csv('raw/train_net.csv')

cols_to_drop_train = ['FLOW_ID', 'ID', 'ALERT', 'ANOMALY', "ANALYSIS_TIMESTAMP"]  

colums_to_drop_train = []

for c in cols_to_drop_train:
    if c in df_train.columns:
        colums_to_drop_train.append(c)

X_train = df_train.drop(columns=colums_to_drop_train).copy()
X_train.head()
```

```python
categorical_columns = X_train.select_dtypes(include=['object']).columns
```

```python
categorical_columns
```

```python
for column in categorical_columns:
    le = LabelEncoder()
    X_train[column] = le.fit_transform(X_train[column])
X_train.head()
```

```python
df_test= pd.read_csv('raw/test_net.csv')
```

```python
df_test.head()
```

```python
data_robust_scaler = RobustScaler().fit_transform(X_train)
data_robust_scaler
```

```python
# TEST
cols_to_drop_test = ['FLOW_ID', 'ID', "ANALYSIS_TIMESTAMP"]
colums_to_drop_test = []
for c in cols_to_drop_test:
    if c in df_test.columns:
        colums_to_drop_test.append(c)
Y_test = df_test.drop(columns=colums_to_drop_test).copy()

Y_test.head()
```

```python
cat_cols_test = Y_test.select_dtypes(include= ["object"]).columns
cat_cols_test
```

```python
for c in cat_cols_test: 
    le = LabelEncoder()
    Y_test[c] = le.fit_transform(Y_test[c]) 
Y_test.head()
```
